```python
import numpy as np
```

```python
scores_path = "./iso_forest/scores"
pred_path = "./iso_forest/predictions"

mnist_scores_fc1 = np.load(f"{scores_path}/mnist-test-fc1.npy")
mnist_predictions_fc1 = np.load(f"{pred_path}/mnist-test-fc1.npy")
notmnist_scores_fc1 = np.load(f"{scores_path}/notmnist-fc1.npy")
notmnist_predictions_fc1 = np.load(f"{pred_path}/mnist-test-fc1.npy")
```

```python
print(mnist_scores_fc1[0:5])
print(np.mean(mnist_scores_fc1))
print(np.max(mnist_scores_fc1))
print(np.min(mnist_scores_fc1))

print(notmnist_scores_fc1[0:5])
print(np.mean(notmnist_scores_fc1))
print(np.max(notmnist_scores_fc1))
print(np.min(notmnist_scores_fc1))
```

```python
import matplotlib.pyplot as plt

range_max = -0.3
range_min = -0.7
num_bins = 100
bins = np.linspace(range_min, range_max, num_bins)
plt.hist(mnist_scores_fc1, bins, alpha=0.5, label="MNIST anomaly score")
plt.hist(notmnist_scores_fc1, bins, alpha=0.5, label="notMNIST anomaly score")
plt.legend(loc='upper right')
plt.title("Anomaly score of fc1 layer")
plt.xlabel(f"anomaly scores (bin size={(range_max-range_min)/num_bins})")
plt.ylabel("number of samples")
plt.show()
```
